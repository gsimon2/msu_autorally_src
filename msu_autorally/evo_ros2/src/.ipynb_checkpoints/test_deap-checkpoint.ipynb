```python
%matplotlib inline
import zmq
import pandas as pd
from sklearn.metrics import mean_squared_error


import random
from deap import creator, base, tools, algorithms


### Set up individual's shape and fitness function ###
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.FitnessMin)


GENOME_SIZE = 4

toolbox = base.Toolbox()
toolbox.register("attr_float", random.random)
toolbox.register("individual", tools.initRepeat, creator.Individual,
                 toolbox.attr_float, n=GENOME_SIZE)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)


### Set up communication sockets ###
IP_ADDR = '127.0.0.1'
SEND_PORT = 5022
RECV_PORT = 5032

try:
    #Initialize the socket for data
    # Setup the socket to send data out on.
    context = zmq.Context()
    socket = context.socket(zmq.PUSH)
    #socket.setsockopt(zmq.LINGER, 0)    # discard unsent messages on close
    socket.bind('tcp://{}:{}'.format(IP_ADDR, SEND_PORT))

    # Setup the socket to read the responses on.
    receiver = context.socket(zmq.PULL)
    receiver.bind('tcp://{}:{}'.format(IP_ADDR, RECV_PORT))

    # Setup ZMQ poller
    poller = zmq.Poller()
    poller.register(receiver, zmq.POLLIN)
    

    ### Set up evaluation function ###
    def evaluateInd(ind):
        print('Sending ind: {}'.format(ind))
        socket.send_json(ind)
        
        print('Waiting Result')
        result = receiver.recv_json()
        
        print('Recv\'d Result')
        df = pd.DataFrame.from_dict(dict(result))
        df['error'] = abs(df['Actual Speed'] - df['Goal Speed'])
        df.plot(x='Time')
        fitness = mean_squared_error(df['Actual Speed'],  df['Goal Speed'])
        print('Fitness: {}'.format(fitness))
        
        return (fitness, )
        


    ### Set up Evo Algo operators ###
    toolbox.register("mate", tools.cxTwoPoint)
    toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=0.2)
    toolbox.register("select", tools.selTournament, tournsize=3)
    toolbox.register("evaluate", evaluateInd)


    ### GA ###
    pop_size = 3
    number_generations = 3
    population = toolbox.population(n=pop_size)


    for gen in range(number_generations):
        offspring = algorithms.varAnd(population, toolbox, cxpb=0.5, mutpb=0.1)
        fits = toolbox.map(toolbox.evaluate, offspring)
        for fit, ind in zip(fits, offspring):
            ind.fitness.values = fit
        population = toolbox.select(offspring, k=len(population))


    top = tools.selBest(population, k=2)
    print(top)
    print(population[0].fitness)
    

finally:
    socket.close()
    receiver.close()
    context.destroy()
```
