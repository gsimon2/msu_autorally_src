```python
%matplotlib inline
import zmq
import pandas as pd

from sklearn.metrics import mean_squared_error

IP_ADDR = '127.0.0.1'
SEND_PORT = 5022
RECV_PORT = 5032

try:
    #Initialize the socket for data
    # Setup the socket to send data out on.
    context = zmq.Context()
    socket = context.socket(zmq.PUSH)
    #socket.setsockopt(zmq.LINGER, 0)    # discard unsent messages on close
    socket.bind('tcp://{}:{}'.format(IP_ADDR, SEND_PORT))

    # Setup the socket to read the responses on.
    receiver = context.socket(zmq.PULL)
    receiver.bind('tcp://{}:{}'.format(IP_ADDR, RECV_PORT))

    # Setup ZMQ poller
    poller = zmq.Poller()
    poller.register(receiver, zmq.POLLIN)

    genome = list()
    genome = [1,2,3,4]
    #genome = 'end'


    print('Sending addr: {}'.format('tcp://{}:{}'.format(IP_ADDR, SEND_PORT)))
    
    for i in range(1):
        print('here')
        socket.send_json(genome)
        print('here')
        result = receiver.recv_json()
        print result
        df = pd.DataFrame.from_dict(dict(result))
        #df = df[['Time', 'Goal Speed', 'Actual Speed']]

finally:
    socket.close()
    receiver.close()
    context.destroy()
```

```python
df['error'] = abs(df['Actual Speed'] - df['Goal Speed'])
print(mean_squared_error(df['Actual Speed'],  df['Goal Speed']))
df.plot(x='Time')
```
